Evolution of the CO2 concentration in the atmosphere.

```python
# | column: page
import pandas as pd
import altair as alt

df = pd.read_parquet("../data/climate_co2_global_trend.parquet")
df["date"] = pd.to_datetime(df["date"])

# Altair interactive line plot

c = (
    alt.Chart(df)
    .mark_line()
    .encode(
        x="date:T",
        y=alt.Y("trend:Q", scale=alt.Scale(domain=[390, 430]), title="CO2 (ppm)"),
        tooltip=["date:T", "trend:Q"],
    )
    .properties(width="container", height=200)
    .configure_view(strokeWidth=0)
    .configure_axis(grid=False, labelFontSize=14, titleFontSize=14)
    .configure_title(fontSize=16, anchor="start", color="gray")
    .display()
)
```
